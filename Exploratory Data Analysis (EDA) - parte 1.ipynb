## Estatística Univariada, Bivariada e Multivariada


### 1. Dados univariados

Este tipo de dados consiste em apenas uma variável . A análise de dados univariados é, portanto, a forma mais simples de análise, uma vez que a informação trata de apenas uma quantidade que muda. Não lida com causas ou relações e o objetivo principal da análise é descrever os dados e encontrar os padrões que existem dentro deles. O exemplo de dados univariados pode ser a altura.

"A Estatística Univariada inclui todos os métodos de Estatística Descritiva que permitem a análise de cada variável separadamente e também métodos de Estatística Inferencial para determinada variável, podendo esta ser medida para uma ou mais amostras independentes. A análise de variância simples (e o teste "t", em particular) é o exemplo típico de um método de Estatística Univariada (Reis, 1997), pois a palavra «univariada» implica que há uma só variável dependente (Burns, 2000; M. Hill & A. Hill, 2000; Thomas & Nelson, 1996). Sin.: Estatística Univariável (Dorsch et al., 2001); Estatística a uma dimensão (Dagnelie, n.d.)."

### 2. Dados bivariados

Este tipo de dados envolve duas variáveis diferentes . A análise deste tipo de dados trata de causas e relações e a análise é feita para descobrir a relação entre as duas variáveis. Exemplo de dados bivariados podem ser temperatura e vendas de sorvete no verão.

"A Estatística Bivariada inclui métodos de análise de duas variáveis, podendo ser ou não estabelecida uma relação de causa/efeito entre elas. São exemplos típicos de métodos de análise bivariada o teste para a independência de duas variáveis (vulgarmente conhecido por teste do c 2) e o estudo da relação linear entre duas variáveis, quer através dos coeficientes de correlação linear de Pearson ou Spearman, quer do modelo clássico de regressão linear simples (Reis, 1997). Sin.: Estatística Bivariável (Dorsch et al., 2001); Estatística a duas dimensões (Dagnelie, n.d.)."

<img src="images/example-bivariate.png" width="50%">

### 3. Dados multivariados

Quando os dados envolvem três ou mais variáveis , são categorizados em multivariados. Um exemplo desse tipo de dado é supor que um anunciante deseja comparar a popularidade de quatro anúncios em um site, então suas taxas de cliques podem ser medidas para homens e mulheres e as relações entre as variáveis podem então ser examinadas.

"A Estatística Multivariada inclui os métodos de análise das relações de múltiplas variáveis dependentes e/ou múltiplas variáveis independentes, quer se estabeleçam ou não relações de causa/efeito entre estes dois grupos. São também incluídos na Estatística Multivariada os métodos de análise das relações entre indivíduos caracterizados por duas ou mais variáveis. Só os métodos de Estatística Multivariada permitem que se explore a performance conjunta das variáveis e se determine a influência ou importância de cada uma, estando as restantes presentes (Reis, 1997). Sin.: Estatística Multivariável (Dorsch et al., 2001); Estatística a três ou mais dimensões (Dagnelie, n.d.)."

## Variáveis categóricas (Categorical data)

As variáveis categóricas contêm um número finito de categorias ou grupos distintos. Os dados categóricos podem não ter uma ordem lógica. Por exemplo, os preditores categóricos incluem gênero, tipo de material e método de pagamento.

### Nominal

As nominais é quando as categorias não possam ser ordenadas de alguma maneira. Ex.: nomes, cores, sexo.

* Os valores representam unidades discretas
* Mudando as ordens das unidades, não altera os valores

#### Dados Nominais
* Frequência 
    * Conta o número de eventos de interesse
    

* Proporção
    * Divide a sequência pelo número total de eventos
    
    
* Percentagem
    * Multiplica a proporção por 100
    
    
* Ilustra com gráfico de barra ou pizza


### Ordinal

As ordinais podem ser ordenadas. Ex.: tamanho (pequeno, médio, grande), classe social (baixa, média, alta), grau de instrução (básico, médio, graduação, pós-graduação).

* Os valores representam unidades discretas e ordenadas
* A distância entre unidades não é a mesma


#### Dados Ordinais

* Resumo 
   * Frequência, proporção e porcentagens
   * Percentil
   * Moda
   * Mediana
   * Range Interquartil
    
    
* Ilustra com gráfico de barra ou pizza

#### Interval

* unidades ordenadas com valores intermediários
* A distância entre unidades é a mesma
* Sem zero absoluto
    - Origem arbitrária
    - Uma pessoa com um QI de 160 NÃO é 2x mais inteligente de alguém com um QI de 80
        

#### Ratio

* unidades ordenadas com valores intermediários
* A distância entre unidades é a mesma
* Sem zero absoluto
    - Origem é o zero

## Variável numéricas 

### Variáveis contínuas

Variáveis contínuas são variáveis numéricas que têm um número infinito de valores entre dois valores quaisquer. Uma variável contínua pode ser numérica ou de data/hora. Por exemplo, o comprimento de uma peça ou a data e hora em que um pagamento é recebido.
    
### Variáveis Discretas

As variáveis discretas são características mensuráveis que podem assumir apenas um número finito ou infinito contável de valores e, assim, somente fazem sentido valores inteiros. Geralmente são o resultado de contagens. Exemplos: número de filhos, número de bactérias por litro de leite, número de cigarros fumados por dia.

<br>

* Resumo
    * Percentil, mediana, range interquartile
    * Média, mediana ou moda
    * Desvio padrão, range ou IQR


* Ilustração
    * Histograma
    * Diagrama de caixa   
    

## Estatística Descritiva

* Tendência Central
    * Média
    * Mediana
    * Moda
----
* Variabilidade
    * Range, max, min
    * Desvio Padrão
    * Range interquartil, IQR = Q3 - Q1
    

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
credit = pd.read_csv("datasets/credit_card_approvals_clean.csv")
```

```python
credit.tail()
```

<p style="font-size: 18px"><strong>Colunas com valores catégoricos:</strong> Gender, Married, BankCustomer, Industry, Ethnicity, PriorDefault, Employed, DriversLicense, Citizen, Approved</p>
<p style="font-size: 18px"><strong>Colunas com valores contínuos:</strong> Age, Debt, YearsEmployed, CreditScore, Income</p>


```python
# Vou deletar ZipCode, pois não será usado
credit = credit.drop(columns='ZipCode')
credit.info()
```

## Análise Univariada de Variáveis Contínuas

```python
credit[["Age", "Debt", "YearsEmployed", "CreditScore", "Income"]].describe()
```

```python
# Pela análise desse gráfico, percebe-se que as pessoas não solicitam cartão de crédito depois dos 50 anos. No entanto, 
# entre 20 e 40 anos eles solicitam mais comparado as outras faixas etárias.

sns.histplot(credit['Age'], kde=True)
```

```python
# Aqui mostra que as pessoas tendem a fazer cartão de crédito no começo de suas carreiras com YearsEmployed < 10, enquanto no
# YearsEmployed > 10  eles possuem cartão de crédito quando tem experiência profissional. Assim, eles não precisam solicitar
# cartões nessa fase

sns.histplot(credit['YearsEmployed'], kde=True, bins=7)
```

## Análise Univariada de Variáveis Categóricas

```python
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(figsize=(8,6), nrows=2, ncols=2)

# Masculino (Gender -1 ) se inscreveram mais do que o Feminino (Gender -0)
sns.countplot(x='Gender', data=credit, ax=ax1) 

# Pessoas com contas bancárias se inscreveram mais do que pessoas que não têm contas bancárias. Isso não causa surpresa.
sns.countplot(x='BankCustomer', data=credit, ax=ax2) 

# A tendência da etnia pode ser devido a região que os dados foram coletados.
sns.countplot(x='Ethnicity', data=credit, ax=ax3)

# A maioria das inscrições foram rejeitadas
sns.countplot(x='Approved', data=credit, ax=ax4)

plt.tight_layout()
plt.show()
```

## Análise Bivariada de Variáveis Contínuas

```python
credit[["Age", "Debt", "YearsEmployed", "CreditScore", "Income"]].corr()
```

```python
sns.scatterplot(x=credit['YearsEmployed'], y=credit['Income'], data=credit)
plt.ylim(0, 20000)
```

```python
# As empresas não emitem cartões de crédito para pessoas com baixa pontuação de crédito e baixa renda. 
# Além disso, as empresas preferem histórico de emprego decente para a emissão de cartões de crédito.

credit.groupby('Approved').agg('mean')[["Age", "Debt", "YearsEmployed", "CreditScore", "Income"]]
```

```python
fig, axs = plt.subplots(figsize=(13,5), ncols=3)

sns.kdeplot(data=credit, x='Age', hue='Approved', fill=True, ax=axs[0])
sns.kdeplot(data=credit, x='CreditScore', hue='Approved', fill=True, ax=axs[1])
sns.kdeplot(data=credit, x='YearsEmployed', hue='Approved', fill=True, ax=axs[2])

plt.tight_layout()
plt.show()
```

## Análise Bivariada de Variáveis Categóricas vs Variáveis Categóricas

```python
sns.countplot(data=credit, x='Approved', hue='Gender')
```

```python
# Na tabela acima, podemos ver que o percentual de aceitação para ambos os sexos é muito próximo (53% está próximo de 56,4%). 
# Portanto, parece que não houve discriminação de nenhum gênero.

a = pd.crosstab(credit['Gender'], credit['Approved'], margins=True)['All']
pd.crosstab(credit['Gender'], credit['Approved']).divide(a, axis=0).dropna()
```

```python
fig, axs = plt.subplots(figsize=(13,5), ncols=3)

sns.countplot(data=credit, x='Ethnicity', hue='Approved', fill=True, ax=axs[0])
sns.countplot(data=credit, x='Employed', hue='Approved', fill=True, ax=axs[1])
sns.countplot(data=credit, x='BankCustomer', hue='Approved', fill=True, ax=axs[2])

plt.tight_layout()
plt.show()

# No primeiro gráfico de contagem dos três acima, vemos que, para a etnia 'latina', a maioria dos pedidos foi rejeitada. 
# Então houve alguma discriminação contra eles? Vamos tentar descobrir.
```

```python
credit.groupby('Approved').agg('mean')[["Age", "Debt", "YearsEmployed", "CreditScore", "Income"]]
```

```python
# Agora, se compararmos o CreditScore médio da etnia latina (1,85) com o CreditScore médio de todas 
# as solicitações aprovadas (4,60), descobrimos que o Latino teve menos CreditScore do que a população 
# com aplicativos aprovados. O mesmo pode ser visto na coluna Renda, onde o Latino tem média. renda de 434,64 
# onde os pedidos aprovados têm avg. renda superior a 2000. Por esta observação, podemos dizer que é muito improvável 
# que houvesse qualquer discriminação contra o grupo latino.

credit[credit['Ethnicity'] == 'Latino'][["Age", "Debt", "YearsEmployed", "CreditScore", "Income"]].agg('mean')
```

## Análise multivariada

A análise multivariada é uma forma mais complexa de técnica de análise estatística e é usada quando há mais de duas variáveis no conjunto de dados.

Aqui, tentaremos ver as relações entre as variáveis contínuas e a coluna ‘Aprovado’. Para fazer isso, vamos traçar um gráfico de pares, com Hue como aprovado

```python
sns.pairplot(data=credit[["Age", "Debt", "YearsEmployed", "CreditScore", "Income", 'Approved']], hue='Approved', height=2)
```
